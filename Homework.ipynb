## In-Place Algorithms

#### Syntax

```python
#var[i], var[j] = var[j], var[i]
# sometimes known as a swap algorithm
# we sometimes use this as a helper function, seen later in the doc

def swap(alist, x, y, z):
    alist[x], alist[y], alist[z] = alist[z], alist[y], alist[x]
    return alist

my_list = [20, 4, 10]
print(f"before swap: {my_list}")
swap(my_list, 2, 0, 1)
print(f"After swap: {my_list}")
```

## Two Pointers

#### Syntax

```python
# alist[left], alist[right], = alist[right], alist[left]
# use two pointers to swap, you can use a while loop in most cases

my_list = [1,2,3,12,9,8,4,11,22]
def twoPointers(alist):
    #creating pointers for the list
    left = 0 #left is index 0, starting at the start of the list (or the left side...)
    right = len(alist) - 1 #the last index (or the right side...)
    #These are the "two pointers" they will meet each other in the middle, breaking the loop
    while left <= right:
        alist[left], alist[right] = alist[right], alist[left]
        left += 1
        right -= 1
    return alist

twoPointers(my_list)
    
    
```

# Sorting Algorithms

```python
# O(1) aka constant time
array = [1,2,3,4]
print(array[0])

# O(n) aka linear time
# because it has to check each index, scales linearly with length of the list
if 2 in array:
    print("its here")
for num in array:
    print(num)

#O(n Logn)
array2 = [1, 5, 6, 8, 4, 6, 7, 4, 5, 2, 4]
sorted_array = sorted(array2)
array2.sort()
print(sorted_array)
print(array2)

#O(n^2) aka Quadratic
# basically any nested linear operation will exponentially increase the time complexity
# ex: double for loops are quadratic or nested ifs in for loops
array = [1,2,3,4]
array2 = [1,5,6,8,4,6,7,4,5,2,4]
for x in array:
    if x in array2:
        print(x)
for x in array2:
    for y in array2:
        print(x,y)

#cubic aka exponential
for x in array2:
    for y in array2:
        for z in array2:
            print(x,y,z)
```

#### Bubble Sort

Worst Case: O(n^2) Time - O(1) Space

```python
#Time Complexity
# How many actions have to happen per things taken inputs

#.count() is linear
#.append() is constant
#.sort()/sorted() O(n logn)
#.pop() is constant
#.pop(x) is linear because it has to traverse the list to remove the given index
```

```python
#Best case O(n) (linear)
#Worst case O(n^2) (Quadratic)

#after each pass, one number will reach its final position

my_list = [22, 55, 88, 44, 1, 100, 34, 66]
def swap(i, j, array):
    array[i], array[j] = array[j], array[i]
    
def bubbleSort(array):
    isSorted = False #Setting a flag here
    while not isSorted:
        isSorted = True
        for index in range(len(array)-1):
            if array[index] > array[index + 1]:
                swap(index, index + 1, array)
                isSorted = False
    return array
bubbleSort(my_list)
```

##### Insertion Sort

Worst Case: O(n^2) time - O(1)space

```python
def swap(i, j, array):
    array[i], array[j] = array[j], array[i]
    
def insertionSort(array):
    for i in range(1,len(array)): #consider the 0 index to be sorted
        #creating a check so we don't go past the 0 index
        j = i
        while j > 0 and array[j] < array[j-1]:
            swap(j, j-1, array)
            j -= 1
    return array

my_list = [22, 55, 88, 44, 1, 100, 34, 66]

insertionSort(my_list)
            
```

## Merge Sort

#### How it Works

```python
#Step 1: split everything into its own group
#Step 2: From left to right, merge two groups together
#Step 3: While merging, place each item in the correct position within the merged group
#Step 4: Continue until only one group is left
#time complexity is O(n logn)
from random import randint
#generating a random list of numbers
nums = [randint(0,20) for i in range(5)]

print(nums)

#merge sort function
def mergeSort(alist):
    print("Splitting.....", alist)
    
    #Step 1
    if len(alist) > 1:
        mid = len(alist) // 2  #I believe this notation divides then floors the result
        lefthalf = alist[:mid]
        righthalf = alist[mid:]
        
        #Recursively call mergeSort to perform splits util len(alist) == 1
        #merge once the splits are done
        mergeSort(lefthalf)
        mergeSort(righthalf)
        
        #index pointers for our list
        l = 0 #pointer for left half
        r = 0 #pointer for right half
        m = 0 #pointer for our main array
        
        #Step 2: compare lefthalf with righthalf
        while l < len(lefthalf) and r < len(righthalf):
            if lefthalf[l] < righthalf[r]:
                alist[m] = lefthalf[l]
                l += 1
            else:
                alist[m] = righthalf[r]
                r+= 1
            m += 1
        #Step 3
        while l < len(lefthalf):
            alist[m] = lefthalf[l]
            l += 1
            m += 1
        while r<len(righthalf):
            alist[m] = righthalf[r]
            r += 1
            m +=1
    
    print("Merging: ", alist)
    return alist

mergeSort(nums)
        
```

# Binary Search

The Binary Search algorithm works by finding the number in the middle of a given array and comparing it to the target. Given that the array is sorted

* The worst case run time for this algorithm is `O(log(n))`

```python
#Less == left side
#Greater == right side
#list of numbers must be sorted in order for this to work

def bin_helper(array, target, left, right):
    """helps us move through the list"""
    while left <= right:
        middle = (left + right) // 2 #adding then flooring
        potentialMatch = array[middle]
        if target == potentialMatch:
            return f"The index is.... {middle}"
        elif target < potentialMatch:
            right = middle - 1
        else:
            left = middle + 1
    return -1

def binarySearch(array, target):
    return bin_helper(array, target, 0, len(array) - 1)

binarySearch([22, 44, 55, 66, 88, 100], 44)
```

Write a program to implement a Linear Search Algorithm. Also in a comment, write the Time Complexity of the following algorithm.

#### Hint: Linear Searching will require searching a list for a given number. 

```python
nums = [12, 64, 15, 43, 84, 73, 67, 36, 27]
# time complexity is O(n) - linear
def search_num(nums, target):
    for i in range(len(nums)):
        if nums[i] == target:
            return f"{target} found at index {i}."
    return "Not in list"

search_num(nums, 67)
```
